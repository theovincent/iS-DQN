```python
from slimdqn.networks.dqn import DQN
from slimdqn.networks.isdqn import iSDQN
import jax

q_dqn = DQN(jax.random.PRNGKey(0), (84, 84, 4), 10, [16, 32, 32, 512], True, "impala", 0, 0, 1, 1, 1)
print("TD-DQN", sum(x.size for x in jax.tree.leaves(q_dqn.params)) * 2)
print("TF-DQN", sum(x.size for x in jax.tree.leaves(q_dqn.params)))

for K in [1, 4, 9, 49]:
	q_isdqn = iSDQN(jax.random.PRNGKey(0), (84, 84, 4), 10, K, [32, 64, 64, 512], True, False, "impala", 0, 0, 1, 1, 1, 1)

	print(f"iSDQN K={K}", sum(x.size for x in jax.tree.leaves(q_isdqn.params)))
```
